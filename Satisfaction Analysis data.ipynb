# Descriptive Statistics - Participant's Demographics

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Survey_data_final.csv')

# 1. Convert respondent_age into categorical variables (age groups)
bins = [0, 18, 35, 50, 65, 100]  # Age group bins (0-18: Children, 19-35: Young Adults, etc.)
labels = ['Children (0-18)', 'Young Adults (19-35)', 'Middle-aged (36-50)', 'Older Adults (51-65)', 'Seniors (65+)']
df['age_group'] = pd.cut(df['respondent_age'], bins=bins, labels=labels, right=False)

# Descriptive statistics by district for numeric variables (hh_size_adults, hh_size_children)
descriptive_stats = df.groupby('district')[['hh_size_adults', 'hh_size_children']].describe()
print(descriptive_stats)

# Count and proportion by district for categorical variables (respondent_gender, marital_status, education_level)
# respondent_gender distribution
gender_distribution = df.groupby('district')['respondent_gender'].value_counts(normalize=True).unstack()
print(gender_distribution)

# marital_status distribution
marital_status_distribution = df.groupby('district')['marital_status'].value_counts(normalize=True).unstack()
print(marital_status_distribution)

# education_level distribution
education_distribution = df.groupby('district')['education_level'].value_counts(normalize=True).unstack()
print(education_distribution)

# Function to add labels on top of each bar
def add_labels(ax, is_count=True):
    for p in ax.patches:
        label = int(p.get_height()) if is_count else format(p.get_height(), '.2f')
        ax.annotate(label,
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    xytext=(0, 9), textcoords='offset points')

# 1. Bar plot for hh_size_adults by district with labels (decimals)
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='district', y='hh_size_adults', data=df, ci=None)
plt.title('Average Household Size (Adults) by District')
plt.xticks(rotation=45)
add_labels(ax, is_count=False)  # Keep the average household size as decimals
plt.show()

# 2. Bar plot for hh_size_children by district with labels (decimals)
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='district', y='hh_size_children', data=df, ci=None)
plt.title('Average Household Size (Children) by District')
plt.xticks(rotation=45)
add_labels(ax, is_count=False)  # Keep the average household size as decimals
plt.show()

# 3. Bar plot for age_group distribution by district with labels (integers)
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='district', hue='age_group', data=df)
plt.title('Age Group Distribution by District')
plt.xticks(rotation=45)
add_labels(ax, is_count=True)  # Display counts as integers
plt.show()

# 4. Count plot for respondent_gender by district with labels (integers)
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='district', hue='respondent_gender', data=df)
plt.title('Respondent Gender Distribution by District')
plt.xticks(rotation=45)
add_labels(ax, is_count=True)  # Display counts as integers
plt.show()

# 5. Count plot for marital_status by district with labels (integers)
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='district', hue='marital_status', data=df)
plt.title('Marital Status Distribution by District')
plt.xticks(rotation=45)
add_labels(ax, is_count=True)  # Display counts as integers
plt.show()

# 6. Count plot for education_level by district with labels (integers)
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='district', hue='education_level', data=df)
plt.title('Education Level Distribution by District')
plt.xticks(rotation=45)
add_labels(ax, is_count=True)  # Display counts as integers
plt.show()
```

## Visit Frequency by District 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('Survey_data_final.csv')

# Check unique values in visit_frequency
print(df['visit_frequency'].unique())

# Assuming numeric values need to be converted to categories, e.g., 
# mapping numerical visit frequencies to more understandable strings
visit_freq_map = {
    1: 'Rarely',
    2: 'Occasionally',
    3: 'Sometimes',
    4: 'Often',
    5: 'Very Often'
}

# Check if visit_frequency is numeric and apply mapping
if df['visit_frequency'].dtype in ['int64', 'float64']:
    df['visit_frequency'] = df['visit_frequency'].map(visit_freq_map)

# Convert to categorical type
df['visit_frequency'] = pd.Categorical(df['visit_frequency'], 
                                        categories=['Rarely', 'Occasionally', 'Sometimes', 'Often', 'Very Often'], 
                                        ordered=True)


# Set the visual style
sns.set(style="whitegrid")

# Create the count plot
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='district', hue='visit_frequency', data=df, palette='Set2')

# Title and labels
plt.title('Visit Frequency by District', fontsize=16)
plt.xlabel('District', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add count labels on top of the bars
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fmt='%d', padding=3)

# Adjust layout for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```

# Satisfaction derived from Healthcare Services across the Study Population

```python
import pandas as pd

# Load the data
df = pd.read_csv('Survey_data_final.csv')
```

```python
# Overview of the dataset
print(df.head())
print(df.dtypes)
```

```python
# Check for missing values
satisfaction_columns = [
    'overall_satisfaction',
    'cleanliness_satisfaction',
    'staff_attitude_satisfaction',
    'medicine_availability_satisfaction',
    'waiting_time_satisfaction',
    'cost_satisfaction',
    'quality_satisfaction',
    'communication_satisfaction',
    'feedback_satisfaction'
]

missing_values = df[satisfaction_columns].isnull().sum()
print("Missing values in satisfaction columns:\n", missing_values)
```

```python
# Convert to numeric and drop NaN values
df[satisfaction_columns] = df[satisfaction_columns].apply(pd.to_numeric, errors='coerce')
df_cleaned = df.dropna(subset=satisfaction_columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visualization style
sns.set(style="whitegrid")

# Boxplots for each satisfaction metric
plt.figure(figsize=(18, 12))
for i, column in enumerate(satisfaction_columns):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(x='district', y=column, data=df_cleaned)
    plt.title(column.replace('_', ' ').title() + ' by District')
    plt.xlabel('District')
    plt.ylabel('Satisfaction Score')
    plt.ylim(1, 5)  # Adjust based on your satisfaction scale

plt.tight_layout()
plt.show()
```

# Average Satisfaction by District

```python
# Average satisfaction scores by district
average_satisfaction = df_cleaned.groupby('district')[satisfaction_columns].mean().reset_index()

# Melt the DataFrame for easier plotting
average_satisfaction_melted = average_satisfaction.melt(id_vars='district', var_name='satisfaction_type', value_name='score')

# Bar plot for average satisfaction scores with improved spacing
plt.figure(figsize=(14, 8))
bar_plot = sns.barplot(x='district', y='score', hue='satisfaction_type', data=average_satisfaction_melted, dodge=True)

# Improve bar spacing
bar_plot.set_xticklabels(bar_plot.get_xticklabels(), rotation=45)
plt.title('Average Satisfaction Scores by District', fontsize=16)
plt.xlabel('District', fontsize=14)
plt.ylabel('Average Satisfaction Score', fontsize=14)
plt.ylim(1, 5)

# Adding data labels to the bars
for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height():.2f}', 
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha='center', va='bottom', 
                      fontsize=10, color='black', 
                      rotation=0, 
                      xytext=(0, 5), 
                      textcoords='offset points')

plt.legend(title='Satisfaction Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Overall Satisfaction - Proportion

```python
# Function to convert scores to percentages
def score_to_percentage(score):
    return ((score - 1) / 4) * 100

# Calculate the average satisfaction for each district and convert to percentage
district_means = df_cleaned.groupby('district')[satisfaction_columns].mean()
district_percentage = district_means.applymap(score_to_percentage)

# Calculate overall average satisfaction across all districts and convert to percentage
overall_mean = df_cleaned[satisfaction_columns].mean()
overall_percentage = overall_mean.apply(score_to_percentage)

# Display results
print("Average Satisfaction by District (Percentage):")
print(district_percentage)
print("\nOverall Average Satisfaction (Percentage):")
print(overall_percentage)
```

## Overall Satisfaction By District

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df_cleaned is your DataFrame containing the data

# Function to convert scores to percentages
def score_to_percentage(score):
    return ((score - 1) / 4) * 100

# Calculate the average overall satisfaction for each district and convert to percentage
overall_satisfaction = df_cleaned.groupby('district')['overall_satisfaction'].mean()
overall_percentage = overall_satisfaction.apply(score_to_percentage).reset_index()

# Rename columns for clarity
overall_percentage.columns = ['District', 'Overall Satisfaction (%)']

# Set up the visualization style
sns.set(style="whitegrid")

# Create a summary bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='District', y='Overall Satisfaction (%)', data=overall_percentage, palette='viridis')

# Title and labels
plt.title('Overall Satisfaction by District', fontsize=16)
plt.xlabel('District', fontsize=14)
plt.ylabel('Overall Satisfaction (%)', fontsize=14)
plt.ylim(0, 100)  # Set limits to 0-100% for better visualization

# Annotate bars with percentage values
for index, value in enumerate(overall_percentage['Overall Satisfaction (%)']):
    plt.text(index, value + 1, f'{value:.1f}%', ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```
